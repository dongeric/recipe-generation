```python
import numpy as np
import pandas as pd
from collections import OrderedDict
from bs4 import BeautifulSoup
import requests
from recipe_scrapers import scrape_me
```

```python
#using query=tomato
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
webpage = requests.get('https://www.bbcgoodfood.com/search/recipes?query=tomato', headers = headers)
soup = BeautifulSoup(webpage.content)

links = soup.find("div",{"class":"main row grid-padding"}).find_all("h3",{"class":"teaser-item__title"})
```

```python
tomato = []
for l in links:
    tomato.append(l.a["href"])
```

```python
tomato
```

```python
titles = []
times = []
ingredients = []
instructions = []
links = []
for t in np.arange(3, 3000000, 3):
    scraper = scrape_me('https://www.bbcgoodfood.com/recipes/' + str(t))
    titles.append(scraper.title())
    times.append(scraper.total_time())
    ingredients.append(scraper.ingredients())
    instructions.append(scraper.instructions())
    links.append(scraper.links())
```

```python
titles[2001]
```

```python
df = pd.DataFrame(OrderedDict({'Title': titles, 'Time': times, 'Ingredients':ingredients, 'Instructions':instructions}))
```
